Implement K-Means clustering/ hierarchical clustering on sales_data_sample.csv dataset.Determine the number of clusters using the elbow method.
Dataset link : https://www.kaggle.com/datasets/kyanyoga/sample-sales-data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./sales_data_sample.csv", encoding='unicode_escape')
```

```python
df
```

```python
df.info()
```

```python
df = df.drop(['ADDRESSLINE1','ADDRESSLINE2','STATE','POSTALCODE','PHONE'],axis = 1)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
import datetime as dt 
snapshot_date = df['ORDERDATE'].max() + dt.timedelta(days=1)
df_rfm = df.groupby(['CUSTOMERNAME']).agg({
    'ORDERDATE' : lambda x : (snapshot_date - x.max()).days,
    'ORDERNUMBER' : 'count',
    'SALES' : 'sum'
})
df_rfm.head()
```

```python
df_rfm.rename(columns={'ORDERDATE':'Recent','ORDERNUMBER':'Frequency'},inplace = True)
```

```python
df_rfm.head()
```

```python
df_rfm['M'] = pd.qcut(df_rfm['SALES'],q=4,labels= range(1,5))
df_rfm['F'] = pd.qcut(df_rfm['Frequency'],q=4,labels= range(4,0,-1))
df_rfm['R'] = pd.qcut(df_rfm['Recent'],q=4,labels= range(1,5))
df_rfm['RMF'] = df_rfm[['R','M','F']].sum(axis = 1)
df_rfm.head()
```

```python
def rfm_level(df):
    if bool(df['RMF'] >= 10):
        return "HIGH value Customer"
    elif bool(df['RMF']<10 and df['RMF']>6):
        return "Mid value Customer"
    else:
        return "Low value Customer"
df_rfm['LEVEL'] = df_rfm.apply(rfm_level,axis = 1)
df_rfm
    
```

```python
data = df_rfm.drop(['M','F','R','RMF','LEVEL'],axis = 1)
```

```python
data_log = np.log(data)
data_log
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(data_log)
data_normal = sc.transform(data_log)
data_normal = pd.DataFrame(data_normal,index = data_log.index, columns = data_log.columns)
data_normal
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

sse = {}

for k in range(1,21):
    kmeans = KMeans(n_clusters=k,random_state=1)
    kmeans.fit(data_normal)
    sse[k] = kmeans.inertia_
```

```python
sse
```

```python
plt.figure(figsize=(20,10))
plt.title("THE ELBOW METHOD")
plt.xlabel("k")
plt.ylabel("SSE")
plt.style.use('ggplot')
# plt.plot(sse.keys(),sse.values())
sns.pointplot(x=sse.keys(),y=sse.values())
plt.text(4.5, 60, "Largest Angle", bbox = dict(facecolor = 'lightgreen', alpha = 0.5))
```

```python
kmeans = KMeans(n_clusters=5,random_state=1)
kmeans.fit(data_normal)
kmeans
```

```python
cluster_labels = kmeans.labels_
data = data.assign(Cluster = cluster_labels)
data
```

```python
data['Cluster1'] = kmeans.fit_predict(data_normal)
data
```
