Implement Gradient Descent Algorithm to find the local minima of a function. For example, find the local minima of the function y=(x+3)Â² starting from the point x=2

```python
current_x = 2
learn_rate = 0.01
precision = 0.000001
previous_step_size = 1
max_iteration = 10000
iteration_counter = 0
f = lambda x: (x+3)**2
```

```python
def df(x):
    return 2*(x+3)
```

```python
listx = []
while previous_step_size>precision and iteration_counter<max_iteration:
    previous_x = current_x
    current_x = current_x - learn_rate*df(previous_x)
    previous_step_size = abs(current_x-previous_x)
    iteration_counter+=1
    print(iteration_counter,") current x : ",current_x)
    listx.append(current_x)

print("The local minimum occurs at ",current_x)
```
